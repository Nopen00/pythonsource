````python
# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경구
# * 

def addd(*args):
    result = 0 
    for i in args:
        result = result + i
    print(result)
````

````python
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result += i

    elif choice == 'mul':
        result =1
        for i in args:
            result *= i
    
    return result
````

````python
# a = 10에서 'a ='를 지우고 호출하세요
print(add_mul("add", 10, 10, 11, 12, 13))
print(add_mul("mul", 10, 10, 11, 12, 13))
````

````python
# ** => 키워드 매개변수 

def print_kwargs(**kwargs):
    print(kwargs)

    print(kwargs.keys())
    for k in kwargs.keys():
        print(k)

    print(kwargs.values())
    for v in kwargs.values():
        print(v)
    
    print(kwargs.items())
    for k, v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name = 'kim', age = 15)
````

````python
# 입력으로 숫자(자연수)를 받아서 자연수가 짝수라면 true, 아니면 false 리턴
def is_even(a):
    if a % 2 == 0:
        return True
    else:
        return False
````

````python
# 평균 구하는 함수
def get_average(*args):
    add = 0
    for i in args:
        add += i
    #return '총합 {add}, 평균{add / len(args)} 입니다'
    print(f'총합 {add}, 평균{add / len(args)} 입니다')
````

````python
print(is_even(1))
print(is_even(2))


"짝수" if is_even(5) else '홀수'


print(get_average(1,2,3))
````

````python
def calculator(num1, num2, op):
    if op == '+':
        return num1+ num2
    elif op == '-':
        return num1-num2
    elif op == '*':
        return num1*num2
    elif op == '/':
        return num1/num2
    elif op == '//':
        return num1//num2
    
print(calculator(10,20,'+'))
print(calculator(10,20,'-'))
print(calculator(10,20,'*'))
print(calculator(10,20,'/'))
print(calculator(10,20,'//'))
````
