#### 문자열 포매팅
 - 문자열 안에 어떤 값을 삽입하는 방법

 | 코드 | 설명 |
 | --- | --- |
 | %s | 문자열(string) |
 | %d | 정수(integer) |
 | %c | 문자 1개(character) |
 | %f | 부동 소수(floating-point) |
 | %o | 8진수 |
 | %x | 16진수 |
 | %% | % 문자 자체 |

 - format() 함수 사용
 - f 문자열 포매팅 (파이썬 3.6 버전부터 지원)

```python
a= 'I eat %d apples' % 3
a
```

```python
# %s 는 어떤 값이든 가능
a= 'I eat %s apples' % 'five'
a
```

```python
a = "Error is %d%%" %98
a
```

```python
a = "I ate %d apples, so I was sick for %s days" % (10, "three")
a
```

```python
# hi 문자열 출력시 10개인 공간 사용
# 오른쪽 정렬
print("%10s" % "hi")
# 왼쪽 정렬
print("%-10sjane" % "hi")

# 0 : 길이에 상관하지 않음
# 소수점 네번째 자리까지만
print ("%0.4f" % 3.141285478)
# 전체 길이 10개인 문자열 공간에서 오른쪽 정렬
print ("%10.4f" % 3.141285478)
```

```python
a= 'I eat {} apples' .format(3)
a
```

```python
a= 'I eat {0} apples, so I was sick for {1} days' .format(3, "three")
a
```

```python
name = '홍길동'
age = 30

# 첫번째 방법
print('나의 이름은 %s 나이는 %d이다' %(name, age))

# 두번째 방법
print('나의 이름은 {} 나이는 {}이다' .format(name, age))

# f문자열 포매팅
print(f'나의 이름은 {name} 나이는 {age}이다')
```
