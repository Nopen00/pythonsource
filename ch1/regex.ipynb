### 정규표현식(정규식)
 - 복자한 문자열을 처리할 때 사용하는 방법
 - 파이썬만의 고유한 문법이 아니라 문자열을 처리하는 모든 곳에서 사용하는 일종의 형식 언어
 

```python
# 주민등록번허 => 뒷자리를 모두 * 문자로 변경

# 파이썬에서 정규식 지원하는 모듈

import re
dir(re)

# re. compile() : 정규식을 작석
p = re.compile('ab*')
print(p.match('a'))
print(p.match('abc'))
print(p.match('dudu'))

# 패턴을 특정 문자열이 포함하고 있는가?
```

| 메서드 | 목적 |
|---|---|
| match | 문자열의 처음부터 정규식과 매치되는지 조사 |
| search | 문자열 전체를 검색하여 정규식과 매치되는지 조사 |
| findall | 정규식과 매치되는 모든 문자열(substring)을 리스트로 리턴 |
| finditer | 정규식과 매치되는 모든 문자열(substring)을 이터레이터 객체로 리턴 |

- 자주 사용하는 문자 표현식

| 정규표현식 | 설명 |
|---|---|
| \d | 숫자와 매치 [0-9] 와 동일한 표현식 |
| \D | 숫자가 아닌 것과 매치 [^0-9] 와 동일한 표현식 |
| \s | 화이트스페이스(스페이스,탭,줄바꿈 등)문자와 매치 [빈칸\t\n\r\f\v] 와 동일한 표현식 |
| \S | 화이트스페이스(스페이스,탭,줄바꿈 등)문자가 아닌 것과 매치 [^빈칸\t\n\r\f\v] 와 동일한 표현식 |
| \w | 문자 + 숫자와 매치 [a-zA-Z0-9_] 와 동일한 표현식 |
| \W | 문자 + 숫자가 아닌 것과 매치 [^a-zA-Z0-9_] 와 동일한 표현식 |

```python
p = re.compile("[0-9]+")
p = re.compile(r"\d+")
print(p.match("012 abc"))

# 원본 문자열에서 0~9 가 아닌 문자 찾기
#p = re.compile("[^0-9]+")
p = re.compile(r"\D+")
print(p.match("012 abc"))

# 공백, 탭, 엔터 문자 찾기
p = re.compile("[ \t\n\r\f\v]+")
p = re.compile(r"\s+")
print(p.search("012 abc"))

# 공백, 탭, 엔터 문자가 아닌 문자 찾기
#p = re.compile("[ \t\n\r\f\v]+")
p = re.compile(r"\S+")
print(p.search("012 abc"))


# 대소문자, 숫자, _ 포함한 문자를 찾기
# p = re.compile("[ a-zA-Z0-9_]+")
p = re.compile(r"\w+")
print(p.search("012 abc"))

# 대소문자, 숫자, _ 포함하지 않은 문자 찾기
# p = re.compile("[ ^a-zA-Z0-9_]+")
p = re.compile(r"\W+")
print(p.search("012 abc"))
```

```python
# re.I : ignorecase

# 영어가 포함되어 있는지?
p = re.compile("[a-z]+",re.I)
print(p.match("python"))
print(p.match("Python"))
print(p.match("PYTHON"))
```

```python
# re.M

data = """python one
life is too short
python two
you need python
python three
"""

# ^ : 시작하는
# python 시작하고 스페이스 문자 하나 영어, 숫자 ,_인 문자 찾아줘
p = re.compile(r"^python\s\w+", re.M)
print(p.findall(data))
```

```python
# | : or

p = re.compile('Crow|Servo')
m = p.match('CrowHolle')

print(m)

m = p.match('ServoHello')
print(m)
```

```python
# () : 그루핑

# ABCABC 문자열이 반복되는지 조사

p = re.compile("[ABC]+")
m = p.search("ABCABCABC OK?")
print(m)
print(m.group())
```

```python
p = re.compile("(ABC)+")
m = p.search("ABCABCABC OK?")
print(m)
print(m.group())
print(m.group(1))
```

```python
p = re.compile(r"(\w+)\s+(\d{3}-\d{4}-\d{4})")
m = p.search("park 010-1234-5678")
print(m)
print(m.group())
print(m.group(1))
print(m.group(2))
```

```python
# 찾아서 바꾸기
# John Smith
# Smith, John

p = re.compile(r"(\w+)\s(\w+)")
m = p.search("John Smith")
print(m)

# re.sub(패턴, 바꿀문자열, 원본문자열)
re.sub(p,r"\2, \1","John Smith")
```

```python
p = re.compile(r"\d+")

print(re.fullmatch(p, "1232abc"))
print(re.fullmatch(p, "1232456"))
print(re.fullmatch(p, "abc123"))
```

```python
# 영문 소문자로 시작하고 영문,숫자 포함된 5~11글자

p = re.compile("[a-z][a-z0-9]{5,11}")
print(re.fullmatch(p, "1232abc"))
print(re.fullmatch(p, "1232456"))
print(re.fullmatch(p, "abc123"))

m = re.fullmatch(p, 'abc123')
print(m.group())
```

```python
# 입력을 받은 값이 숫자인지 확인

# 숫자입니다 or 숫자가 아닙니다 출력
# fullmatch()

p = re.compile(r"\d+")
num = input('숫자')

if re.fullmatch(p, num):
    print('숫자 입니다')
else:
    print('숫자 아니다')
```
