```python
# 주차장

# 사용자에게 메뉴 제공
# 1. 주차,  2. 출차, 3. 종료

# 1
# 자동차 A ~ E
# A 자동차 입차, 주차장 상태 ['A']
# 1
# B 자동차 입차, 주차장 상태 ['A', 'B']
# 2 
# B 자동차 출차, 주차장 상태 ['A']
# 주차장에 5대 초과이면 주차장 만차라고 출력
# 65 == A, 97 == a
num=0
list1 = []

while 1:
    print('1. 주차,  2. 출차, 3. 종료 \n 번호를 선택해 주세요')
    choice = input('1. 주차,  2. 출차, 3. 종료 \n 번호를 선택해 주세요')
    
    if choice == '1':
        print(f'{choice}를 선택하셨습니다')

        if num >= 5:
            print('\n###\t만차 입니다.\t### \n###\t죄송합니다.\t###\n')
            continue

        num+=1
        list1.append(chr(64+num))
        print(f'주차장 상태 ==> [{list1}]')

    elif choice == '2':
        print(f'{choice}를 선택하셨습니다')
        
        if num <= 0:
            print('\n###\t출차할 차가 없습니다.\t###\n###\t     죄송합니다.   \t###\n')
            continue

        num -= 1
        list1.pop() 
        print(f'주차장 상태 ==> [{list1}]')

    elif choice == '3':
        break

    else:
        print('잘못 입력하셨습니다.')
    
    print()
```

```python
print([i for i in range(1,7)])
print([i * i for i in range(5)])
print([i*2 for i in [1,3,5,7,9,16]])
print([[i,i+1,i+2] for i in [1,2,3]])

odd_list = []

for i in range(1,11):
    if i %2 ==1:
        odd_list.append(i)
odd_list

odd_list2 = [i for i in range(1,11) if i %2 ==1]
print(odd_list2)
```

```python
q_list = ['갑', '을', '병', '정', '경']

q_list2 = []
for i in range(5):
    if i != 3:
        q_list2.append(q_list[i])

print(q_list2)


q_list3 = [i for i in q_list if i != '정']
print(q_list3)
```
