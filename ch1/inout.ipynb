### 표준 입출력
 - 표준입력 : 키보드 => input()
 - 표준출력 : 모니터 => print()

```python
num = input('숫자입력')

if num.isdigit():
    num = int(num)+10
    print(num)
```

```python
def calculator(num1, num2, op):
    if op == '+':
        return num1+ num2
    elif op == '-':
        return num1-num2
    elif op == '*':
        return num1*num2
    elif op == '/':
        return num1/num2
    elif op == '//':
        return num1//num2
    
print(calculator(10,20,'+'))
print(calculator(10,20,'-'))
print(calculator(10,20,'*'))
print(calculator(10,20,'/'))
print(calculator(10,20,'//'))

num1 = int(input('숫자를 입력해 주세요.'))
num2 = int(input('숫자를 입력해 주세요.'))
op =  input('연산자를 입력해 주세요.')
calculator(num1,num2,op)
```

```python
values = input('두 숫자와 연산자 기호 입력(ex 15 10 +)')

num1, num2, op = values.split()

calculator(int(num1), int(num2), op)
```

```python
values = input('두 숫자와 연산자 기호 입력(ex 15,10,+)')

num1, num2, op = values.split(',')

calculator(int(num1), int(num2), op)
```

#### flie 읽고 쓰기

| 파일 열기 모드 | 설명 | 
| ------------- | ---- |
| `r` | **읽기(read) 모드** : 파일의 내용을 읽기만 할때 사용. 파일이 존재하지 않는 경우 오류 발생
| `w` | **쓰기(write) 모드** : 파일의 내용을 새로 쓸때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 작성. 없는 경우 새로 생성
| `a` | **추가(append) 모드** : 기존 파일의 마직막에 내용을 추가할 때 사용. 파일이 없는경우 새로 생성
