#### 인스턴스(멤버) 변수
 - 객체마다 따로따로 가지는 데이터

 - self.변수명 형태로 작성

```python
class Car:
    
    def __init__(self, color, speed):
        # self.변수명 형태로 작성합니다.
        self.color = color
        self.speed = speed

    # 동작 (멤버 메소드 예시)
    def upspeed(self, value):
        self.speed += value

    def downspeed(self, value):
        self.speed -= value

    def __str__(self):
        return f'Car(color = {self.color}, speed = {self.speed})'
```

```python
class SelfTest:

    def function1(cls):
        print('function1 called')

    def function2(cls):
        print('function2 called')


test1 = SelfTest()

test1.function1()
test1.function2()
```

```python
class Dog:
    count = 0
    species = '강아지'

    def __init__(self, name, age):
        self.name = name
        self.age = age
        Dog.count += 1

    def bark(self):
        return f'{self.name} : 멍멍'

    @classmethod
    def get_count(cls):
        return f'생성된 {cls.species}의 수는 ( {Dog.count} )대 입니다.'


dog1 = Dog("초코",3)
dog2 = Dog('모찌',7)

print(dog1.name, dog1.age)

print(dog1.bark())
print(dog2.bark())


print(f'강아지 {Dog.count}마리')
print(Dog.get_count())
```

#### 정적 메소드

 - 개체의 상태나 클래스 변수를 번혀 건드리지 않을때 사용

```python
class Math:
    @staticmethod
    def add(a,b):
        return a+b

```

```python
class Car:

    car_count = 0
    
    def __init__(self, color, speed):
        # self.변수명 형태로 작성합니다.
        self.color = color
        self.speed = speed
        Car.car_count += 1

    # 동작 (멤버 메소드 예시)
    def upspeed(self, value):
        self.speed += value

    def downspeed(self, value):
        self.speed -= value

    def __str__(self):
        return f'Car(color = {self.color}, speed = {self.speed})'

    @staticmethod
    def hello():
        print('hello')

car1 = Car("Red", 50)
car2 = Car("white", 70)
car3 = Car("Black", 85)

Car.hello()
print(f'생성된 차량의 수는 ( {Car.car_count} )대 입니다.')
```

| 구 분 | 첫 인자 | 접근 가능 | 주 용도 |
|---|---|---|---|
| `인스턴스 메서드` | `self` | 인스턴스 변수, 클래스 변수 모두 | 객체별 동작 |
| `@classmethod` | `cls` | 클래스 변수만 | 클래스 단위 동작 |
| `@staticmethod` | 없음 | 없음(매개변수로 받은 것만) | 클래스와 관련은 있지만 독립적인 계산 |

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
#   메소드 : deposit(money) : 잔액 = 잔액 + 입금액
#   메소드 : withdraw(money) : 잔액 = 잔액 - 출금액
#   메소드 : showBalance() : 현재 잔액 출력


class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = int(balance)

    def deposit(self, money):
        self.balance += int(money)

    def withdraw(self, money):
        if self.balance >= money:
            self.balance -= int(money)
        else:
            print('잔액이 부족합니다.')
    
    def showBalance(self):
        print(f'{self.owner}님의 잔액은 {self.balance}원 입니다.')

user1 = Account('길동이', 10000)

Account.deposit(user1, 5000)
Account.showBalance(user1)

Account.withdraw(user1, 3000)
Account.showBalance(user1)

Account.withdraw(user1, 30000)
Account.showBalance(user1)
```

```python
# Book 클래스 작성
# 클래스 변수 : book_count
# 멤버 변수 : 제목(title), 저자(author)
#       메소드 : showInfo() => 제목 : 파이썬
#                            저자 : 홍길동
# 현재 책 권수 출력

class Book:
    book_count = 0

    def __init__(self, title, author):
        self.title = title
        self.author = author
        Book.book_count += 1

        

    #@staticmethod
    def showInfo(self):
        print(f'''제목 : {self.title}\n저자 : {self.author}\n현재 책 권수 : {Book.book_count}''')

dum1 = Book('파이썬', '홍길동')

Book.showInfo(dum1)
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 메소드 : getTotalPrice() = 가격 x 수량 : 총 주문금액 반환
#          showOrder() : 출력문 작성
#          메뉴 : 아메리카노
#          가격 : 4000원
#          수량 : 2개
#          총 주문금액 : 8000원

class Order:

    total_order = 0

    def __init__(self, menu, price, count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1
        

    def getTotalPrice(self):
        return self.price * self.count

    def showOrder(self):
        print(f'메뉴 : {self.menu}\n가격 : {self.price}\n수량 : {self.count}\n총 주문금액 : {self.getTotalPrice()}')

dum1 = Order('아아', 2000, 4)
dum2 = Order('뜨아',1500, 3)

dum1.showOrder()
print()
dum2.showOrder()

print(f'전체 주문 수 : {Order.total_order}건')

```

```python
class Parent:
    def __init__(self):
        self.value = '테스트0'
        print('Parent 클래스의 __init__() 호출')

    def test(self):
        print('Parent 클래스의 test() 호출')
        
class Child(Parent):
    pass


child1 = Child()
child1.test()
print(child1.value)
```

```python
class Child(Parent):
    def __init__(self):
        super().__init__()
        print('Parent 클래스의 __init__() 호출')




child1 = Child()
child1.test()
print(child1.value)
```

```python


class AudioVisual:

    def __init__(self, power, volume):
        self.power = power
        self.volume = volume

    def switch(self, on_off):
        self.power = on_off

    def setVolume(self, volume):
        self.volume = volume


class Audio(AudioVisual):

    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        print('lalala'if self.power else 'turn in on')

class TV(AudioVisual):

    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        print('have fun!!'if self.power else 'switch on')

tv = TV(False, 12, 50)
tv.switch(True)
tv.watch()

audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

```python
# Employee 부모 클래스 :
# 멤버 변수 : 이름(name), 급여(pay)
#     메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# 메소드 : getTotalPay() : 급여 + 보너스 출력

class Employee:
    def __init__(self,name, pay):
        self.name = name
        self.pay = pay

    def showInfo(self):
        print(f'{self.name}의 급여는 {self.pay}원 입니다.')

class Manager(Employee):
    def __init__(self, name, pay,bonus):
        super().__init__(name, pay)
        self.bonus = bonus

    def getTooalPay(self):
        print(f'{self.name}의 급여는 총{self.pay + self.bonus}원(기본급 : {self.pay} + 보너스 : {self.bonus}) 입니다.')


man1 = Employee('철수', 500000)
man2 = Employee('영희', 350000)

man1.showInfo()
man2.showInfo()

man1 = Manager('철수', 500000, 70000)
man2 = Manager('영희', 350000, 70000)

man1.getTooalPay()
man2.getTooalPay()

```

```python
# Shape 부모 클래스
# 메서드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : 반지름(radius)
#     메서드 : area() => 원의 넓이 출력

class Shape:
    def area(self):
        print('도형의 넓이')

class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__()
        self.width = width
        self.height = height

    def area(self):
        print(f'직사각형의 넓이는 : {self.width * self.height}입니다.')

class Circle(Shape):
    def __init__(self, radius):
        super().__init__()
        self.radius = radius

    def area(self):
        print(f'원의 넓이는 : {self.radius*self.radius*3.14}입니다.')


a1 = Shape()
a1.area()
a2 = Rectangle(10,5)
a2.area()
a3 = Circle(5)
a3.area()

```
