### 모듈
 - 함수, 변수 또는 클래스를 모아 놓은 파이썬 파일
 - 다른 파이썬 프로그램에서 불러와 사용할수 있도록 만든 파이썬 파일

### 표준 라이브러리
 - 이미 만들어진 유용한 프로그램
 - 표준 라이브러리는 파이썬 설치 시 자동으로 컴퓨터에 설치됨

```python
import math

math.cos(1)
```

```python
# import 모듈명
# from 모듈명 import 사용하고 싶은 함수, 클래스, 변수... => 모듈명 제거 가능
from math import cos, pi, floor, ceil

print(cos(1))
print(ceil(1.3))
print(floor(1.3))
print(pi)
```

```python
# import 모듈명
# from 모듈명 import 사용하고 싶은 함수, 클래스, 변수... => 모듈명 제거 가능
# as
# import 모듈명 as 별칭
# from 모듈명 import 사용하고 싶은 함수 as 별칭

import math as m

print(m.cos(1))
print(m.ceil(1.3))
print(m.floor(1.3))
print(m.pi)
```

```python
from math import sin as s

s(1)
```
