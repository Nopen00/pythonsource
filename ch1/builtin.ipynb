```python
# 학생 5명의 점수를 입력받아 다음 정보를 출력하는 프로그램을 작성하기오

# 출력내용
# 최고점수(max), 최저점수(min), 평균점수(소수 둘째자리)(sum.len), 
# 점수를 오름차순으로 출력(soted), 등수와 함깨 출력(enumerate)

a = [80,22,65,100,77]

print(f'최고점수 = {max(a)}, 최저점수{min(a)}, 평균점수(소수 둘째자리) = {(sum(a)/len(a)):.2f}')
```

```python
# 문자열 입력받아 다음 정보를 출력
# 출력내용 : 입력한 문자, 각 문자의 유니코드 값, 가장 큰 문자, 가장 작은 문자

# ord, chr, max, min, list
# 문자열 입력 : python
# p -> 80
# y-> 121

# 가장 큰 문자 y
# 가장 작은 문자 p

a = 'happy'
alist = []

for i in a:
    alist.append(ord(i))

print(f'입력 : {a}, \n유니코드 : {alist}, \n가장 큰값{max(alist)}, 가장 작은값{min(alist)}')



```

```python
x = [3,7,1,-5,9]
y = [2,-4,8,10,-1]

# zip, abs, enumerate
# 출력
# (3,2) -> 5
a = list(zip(x,y))
list1 = []

for i, name in enumerate(a, start=1):
    # print(i, name)
    # print(type(name[0]))

    list1.append(name[0] + name[1])

print()
print(list1)
```
