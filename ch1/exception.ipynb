```python
chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"

import random

while True:
    try:
        pwd_num = int(input("비밀번호 길이 : "))

        if pwd_num < 4:
            print("비밀번호는 4자리 이상이어야 합니다.")
            continue
        break
    except ValueError:
        print("숫자만 입력하세요")

# 리스트 컴프리헨션
# password = []
# for i in range(pwd_num):
#     password.append(random.choice(chars))
password = [random.choice(chars) for _ in range(pwd_num)]

# print(f"랜덤 비밀번호 : {password}")
# 리스트 => 문자열로 변경
print(f"랜덤 비밀번호 : {''.join(password)}")
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25
# input() => '2001-08-25' => 날짜 형식
import time, datetime

now = datetime.datetime.now()
print(now)

a = input('생년월일을 입력해 주세요 (ex/ 2001-08-25) ')
birth = datetime.datetime.strptime(a,"%Y-%m-%d")



print('현재 나이 %.2f살 입니다.' % float(((now - birth).days)/365))
```

```python
# 다음과 같은 csv 파일이 있다.
# 이름,국어,영어,수학
# 홍길동,90,80,70
# 김철수,90,80,70
# 이영희,90,80,70
# 성춘향,95,80,70

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

# csv.DictReader() 사용, 평균은 둘째 자리까지

import csv

with open("./data/std_sample.csv", "w", encoding="utf-8", newline="") as f:
    headers = ['이름', '국어', '영어', '수학']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 헤더작성
    writer.writeheader()
    writer.writerow({"이름": "홍길동", "국어": 88, "영어": 92, "수학": 70})
    writer.writerow({"이름": "김철수", "국어": 85, "영어": 70, "수학": 80})
    writer.writerow({"이름": "이영희", "국어": 85, "영어": 70, "수학": 80})
    writer.writerow({"이름": "성춘향", "국어": 85, "영어": 70, "수학": 80})




```

```python
with open("./data/std_sample.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    # print(reader)

    for row in reader:
        # print(row) # {'이름': '홍길동', '국어': '88', '영어': '92', '수학': '70'}
        avg = (int(row.get('국어')) + int(row.get('영어')) + int(row.get('수학'))) / 3
        print(f"{row.get('이름')} : {avg:.2f}")
```
